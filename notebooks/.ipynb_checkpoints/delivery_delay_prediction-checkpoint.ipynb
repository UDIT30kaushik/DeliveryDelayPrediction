```python
df["Delivery_person_Age"] = df["Delivery_person_Age"].astype(int)
```

```python
df["Delivery_person_Age"].unique()
```

```python
df["Delivery_person_Age"] = pd.to_numeric(df["Delivery_person_Age"],errors="coerce")
```

```python
df["Delivery_person_Age"].isnull().sum()
```

```python
df["Delivery_person_Age"] = df["Delivery_person_Age"].fillna(df["Delivery_person_Age"].median())
```

```python
df["Delivery_person_Age"].isnull().sum()
```

```python
df["Delivery_person_Age"] = df["Delivery_person_Age"].astype(int)
```

```python
df["Delivery_person_Age"].dtype
```

```python
df["Delivery_person_Ratings"] = pd.to_numeric(df["Delivery_person_Ratings"],errors="coerce")
```

```python
df["Delivery_person_Ratings"].isnull().sum()
```

```python
df["Delivery_person_Ratings"] = (
    df["Delivery_person_Ratings"]
    .fillna(df["Delivery_person_Ratings"].median())
)
```

```python
df["multiple_deliveries"] = pd.to_numeric(
    df["multiple_deliveries"],
    errors="coerce"
)
```

```python
df["multiple_deliveries"].isnull().sum()
```

```python
df["multiple_deliveries"] = (
    df["multiple_deliveries"]
    .fillna(df["multiple_deliveries"].median())
)
```

```python
df["multiple_deliveries"] = (
    df["multiple_deliveries"]
    .astype(int)
)
```

```python
df["Time_taken(min)"] = (
    df["Time_taken(min)"]
    .str.replace("(min) ", "", regex=False)
)
```

```python
df["Time_taken(min)"] = (
    pd.to_numeric(
        df["Time_taken(min)"],
        errors="coerce"
    )
)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(8,5))

sns.histplot(df["Time_taken(min)"], bins=20)

plt.title("Distribution of Delivery Times")
plt.xlabel("Delivery Time (minutes)")
plt.ylabel("Number of Deliveries")

plt.show()
```

```python
df["Time_taken(min)"].mean()
```

```python
df["Time_taken(min)"].median()
```

```python
df["Time_taken(min)"].skew()
```

Delivery times are approximately normally distributed with a slight positive skew (0.485). Most deliveries occur around the average delivery time, while a smaller number of unusually long deliveries create a right tail in the distribution.

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x="Road_traffic_density",
    y="Time_taken(min)",
    data=df
)

plt.show()
```

Analysis of road traffic density shows a positive relationship with delivery time. Deliveries made during traffic jams have the highest median delivery time (approx. 31 minutes), while deliveries under low traffic conditions have the lowest median delivery time ( approx. 20 minutes). This suggests that traffic congestion is a major contributor to delivery delays.
